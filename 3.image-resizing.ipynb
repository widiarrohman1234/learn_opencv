# Image Resizing using OpenCV | Python

Syntax: `cv2.resize(source, dsize, dest, fx, fy, interpolation)`

Parameters:

- source: Input Image array (Single-channel, 8-bit or floating-point) 
- dsize: Size of the output array
- dest: Output array (Similar to the dimensions and type of Input image array) [optional]
- fx: Scale factor along the horizontal axis  [optional]
- fy: Scale factor along the vertical axis  [optional]
- interpolation: One of the above interpolation methods  [optional]

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread(r"image\road.jpg", 1)
cv2.imshow("Original", image)


half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
cv2.imshow("half", half)

bigger = cv2.resize(image, (1050, 1610))
cv2.imshow("bigger", bigger)

stretch_near = cv2.resize(image, (780, 540), interpolation = cv2.INTER_LINEAR)
cv2.imshow("stretch_near", stretch_near)

Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# erode()

```python
# Reading an image in default mode  
image_gg = cv2.imread(r'image\GG.png') 
image_ab = cv2.imread(r'image\abjad.jpg') 
image_hu = cv2.imread(r'image\human.jpg') 
cv2.imshow("ori gg", image_gg)
cv2.imshow("ori ab", image_ab)
cv2.imshow("ori hu", image_hu)

# Creating kernel 
kernel = np.ones((7,7), np.uint8) 

# Using cv2.erode() method  
image_gg = cv2.erode(image_gg, kernel) 
image_ab = cv2.erode(image_ab, kernel) 
image_hu = cv2.erode(image_hu, kernel) 

# Displaying the image  
cv2.imshow("erode gg", image_gg)
cv2.imshow("erode zb", image_ab)
cv2.imshow("erode hu", image_hu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
