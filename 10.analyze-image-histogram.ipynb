# Python Program to analyze an image using Histogram

#### Histogram Calculation
`cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])`
- `images` :
it is the source image of type uint8 or float32 represented as “[img]”.

- `channels` :
it is the index of channel for which we calculate histogram. For grayscale image, its value is [0] and color image, you can pass [0], [1] or [2] to calculate histogram of blue, green or red channel respectively.

- `mask` :
mask image. To find histogram of full image, it is given as “None”.

- `histSize` :
this represents our BIN count. For full scale, we pass [256].
  
- `ranges`
: this is our RANGE. Normally, it is [0,256]. For example:

```python
# importing required libraries of opencv
import cv2
# importing library for plotting
from matplotlib import pyplot as plt
```

```python
def analHisto(path):
    # reads an input image
    img = cv2.imread(path,0)
    # find frequency of pixels in range 0-255
    histr = cv2.calcHist([img],[0],None,[256],[0,256])
    # show the plotting graph of an image
    plt.plot(histr)
    plt.show()
    plt.title("Image")
    plt.imshow(img)
    # cv2.waitKey(0)
```

```python
path = r'image\mountain.png'
analHisto(path)
```

```python
path = r'image\river.png'
analHisto(path)
```

```python
path = r'image\snow.png'
analHisto(path)
```

### In the second example, it directly finds the histogram and plot it. We need not use calcHist(). See the code below:

```python
# read image grayscale
img = cv2.imread('image\snow.png',0)
# read image RGB
# img = cv2.imread('image\snow.png')

# alternative way to find histogram of an image
plt.hist(img.ravel(),256,[0,256])
plt.show()
plt.title("Image")
plt.imshow(img)
```
