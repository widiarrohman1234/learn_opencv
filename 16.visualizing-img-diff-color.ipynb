# Visualizing image in different color spaces

```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
# reads image as RGB 
img = cv2.imread(r'image\Ganeshji.webp') 
  
# shows the image 
plt.imshow(img) 
```

```python
# Reads image as gray scale 
img2 = cv2.imread(r'image\Ganeshji.webp') 
  
# We can alternatively convert 
# image by using cv2color 
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY) 

# Shows the image 
cv2.imshow('image', img2)  
  
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

```python
# as YCrCb color space 
img3 = cv2.imread(r'image\Ganeshji.webp') 
# Convert to YCrCb color space 
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2YCrCb) 
  
# Shows the image 
cv2.imshow('image', img3)  
  
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

```python
# as HSV color space 
img4 = cv2.imread(r'image\Ganeshji.webp') 
# Convert to YCrCb color space 
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2HSV) 
  
# Shows the image 
cv2.imshow('image', img4)  
  
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

```python
# as LAB color space 
img5 = cv2.imread(r'image\Ganeshji.webp') 
# Convert to YCrCb color space 
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2LAB) 
  
# Shows the image 
cv2.imshow('image', img5)  
  
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

```python
# as EdgeMap 
img6 = cv2.imread(r'image\Ganeshji.webp') 
# Convert to YCrCb color space 
img6 = cv2.cvtColor(img6, cv2.COLOR_BGR2LAB) 
  
# Shows the image 
laplacian = cv2.Laplacian(img6, cv2.CV_64F) 
cv2.imshow('EdgeMap', laplacian)  
  
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

```python
# as EdgeMap 
img7 = cv2.imread(r'image\geeksforgeeks.png') 
  
# plot heat map image 
plt.imshow(img7, cmap ='hot')  
```

```python
# as EdgeMap 
img8 = cv2.imread(r'image\geeksforgeeks.png') 
  
# plot heat map image 
plt.imshow(img8, cmap ='nipy_spectral')  
```
